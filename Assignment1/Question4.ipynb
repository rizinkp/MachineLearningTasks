```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

# Load the dataset
df = pd.read_csv("house_price.csv")

# Check skewness and kurtosis before transformation
original_skewness = skew(df["price_per_sqft"].dropna())
original_kurtosis = kurtosis(df["price_per_sqft"].dropna())
print(f"Skewness (Before Transformation): {original_skewness}")
print(f"Kurtosis (Before Transformation): {original_kurtosis}")
```

```python

# Plot histogram to check normality
plt.figure(figsize=(12, 5))
sns.histplot(df["price_per_sqft"].dropna(), bins=70, kde=True, color="blue")
plt.title("Histogram of Price Per Square Foot (Before Transformation)")
plt.xlabel("Price Per Sq Ft")
plt.ylabel("Frequency")
plt.show()
```

```python
# Perform transformations if needed
df["log_price_per_sqft"] = np.log1p(df["price_per_sqft"])  # Log transformation to handle skewness
df["log_price_per_sqft"]
```

```python
# Check skewness and kurtosis after transformation
transformed_skewness = skew(df["log_price_per_sqft"].dropna())
transformed_kurtosis = kurtosis(df["log_price_per_sqft"].dropna())
print(f"Skewness (After Log Transformation): {transformed_skewness}")
print(f"Kurtosis (After Log Transformation): {transformed_kurtosis}")
```

```python

# Plot histogram after transformation
plt.figure(figsize=(12, 5))
sns.histplot(df["log_price_per_sqft"].dropna(), bins=70, kde=True, color="green")
plt.title("Histogram of Price Per Square Foot (After Log Transformation)")
plt.xlabel("Log Price Per Sq Ft")
plt.ylabel("Frequency")
plt.show()
```
